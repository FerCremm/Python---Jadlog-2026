```python

# Projeto ‚Äì Controle de Transa√ß√µes Financeiras

saldo = 0.0
historico = []  # Lista onde vamos guardar as transa√ß√µes em forma de texto

# while True = loop infinito
# O programa s√≥ vai parar quando a gente usar "break" (no caso 3)
while True:
    # Mostra o menu sempre que volta pro in√≠cio do loop
    print("\n MENU DE OPERA√á√ïES ")
    print("1 - Registrar entrada de valor")
    print("2 - Registrar sa√≠da de valor")
    print("3 - Finalizar o programa")


    # input() sempre retorna texto (string)
    # strip() remove espa√ßos no come√ßo/fim
    menu_txt = input("Op√ß√£o: ").strip()

    # Valida√ß√£o 1: usu√°rio apertou Enter sem digitar nada
    if menu_txt == "":
        print("Voc√™ n√£o digitou nada. Tente novamente.")
        continue  # volta pro in√≠cio do while (mostra o menu de novo)

    # Valida√ß√£o 2: se n√£o for n√∫mero (ex: 'a', '1a', '2.5'), rejeita
    # isdigit() s√≥ aceita d√≠gitos (0-9)
    if menu_txt.isdigit() == False:
        print("Op√ß√£o inv√°lida. Digite 1, 2 ou 3.")
        continue

    # Agora que sabemos que √© n√∫mero, podemos converter para int
    menu = int(menu_txt)

    # match/case = parecido com switch/case (menu de op√ß√µes)
    match menu:

        
        # OP√á√ÉO 1: ENTRADA DE VALOR
        
        case 1:
            # Aqui pode dar erro se a pessoa digitar letras
            # Ent√£o usamos try/except para n√£o quebrar o programa
            try:
                valor = float(input("Valor da ENTRADA: ").strip())
            except ValueError:
                print("Valor inv√°lido. Digite um n√∫mero (ex: 150 ou 150.50).")
                continue  # volta pro menu

            # Regra simples: n√£o aceitar entrada negativa
            if valor < 0:
                print("N√£o √© permitido valor negativo.")
                continue

            # Atualiza o saldo (entrada soma)
            saldo += valor

            # Guarda no hist√≥rico (texto formatado com 2 casas decimais)
            historico.append(f"Entrada: R$ {valor:.2f}")

            # Mostra o saldo atualizado
            print(f"Entrada registrada | Saldo atual: R$ {saldo:.2f}")

        
        # OP√á√ÉO 2: SA√çDA DE VALOR
        
        case 2:
            try:
                valor = float(input("Valor da SA√çDA: ").strip())
            except ValueError:
                print("Valor inv√°lido. Digite um n√∫mero (ex: 80 ou 80.00).")
                continue

            if valor < 0:
                print("N√£o √© permitido valor negativo.")
                continue

            # S√≥ deixa tirar dinheiro se tiver saldo suficiente
            if valor <= saldo:
                saldo -= valor  # sa√≠da subtrai
                historico.append(f"Sa√≠da: R$ {valor:.2f}")
                print(f"Sa√≠da registrada | Saldo atual: R$ {saldo:.2f}")
            else:
                print("Saldo insuficiente. Opera√ß√£o cancelada.")

      
        # OP√á√ÉO 3: FINALIZAR PROGRAMA
     
        case 3:
            print("\nHIST√ìRICO COMPLETO")

            # Se n√£o teve nenhuma transa√ß√£o, hist√≥rico est√° vazio
            if len(historico) == 0:
                print("Nenhuma transa√ß√£o registrada.")
            else:
                # Percorre a lista e imprime cada item
                for item in historico:
                    print("-", item)

            # Mostra saldo final e encerra o loop
            print(f"\nSaldo final dispon√≠vel: R$ {saldo:.2f}")
            break  # sai do while True e o programa termina


        # QUALQUER OUTRA OP√á√ÉO
      
        case _:
            print("Op√ß√£o inv√°lida! Tente novamente.")
```
