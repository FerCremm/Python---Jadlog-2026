```python
# Computar entrega por semana.

soma_entrega = 0

for dia in range(6):
    txt = input(f"{dia+1}º dia | Qnt entregas: ").strip()

    if txt == "":
        print("Você não digitou nada. Vou considerar 0.")
        qnt_entregas = 0
    else:
        qnt_entregas = int(txt)

    soma_entrega = soma_entrega + qnt_entregas

print(soma_entrega)

```

```python
# RODAR COM SENHA E LIMITAÇÃO EM CASA

senha_cadastrada = input("Cadastre uma senha: ")
tentativas = 0

while True:
    senha_entrada = input("Digite sua senha: ")
    tentativas += 1

    if senha_entrada == senha_cadastrada:
        print("Acesso permitido")
        break

    if tentativas == 3:
        print("Tentativas máximas atingidas")
        print("Compareça ao RH")
        break
    
```

```python
# Entrada de valores

saldo = 0
franquias = ['LJ Rio de Janeiro 55', 'LJ São Paulo 23', 'LJ Três Rios 11']
valor_historico = []  # lista para guardar as transações

for franquia in franquias:
    valor = float(input(f'{franquia} | Valor: '))

    valor_historico.append(franquia)
    valor_historico.append(valor)

valor_historico

```

```python
for i, d in enumerate(valor_historico):
    if i % 2 == 1:
        print(d)
```

```python
# Cálculo
soma = 0

for indice, dado in enumerate(valor_historico):
    if indice % 2 == 1:
        print(dado * 0.1)
        # soma = soma + dado * 0.1
        soma += dado * 0.01

print(soma)
```
