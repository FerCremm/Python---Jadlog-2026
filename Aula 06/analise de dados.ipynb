```python
def ler_arquivo(nome_arquivo):
    try:
        df = pd.read_excel(f'dados/{nome_arquivo}.xlsx')
        return df 
    except FileNotFoundError as e:
        print(f'Erro: {e}')

def criar_atualizar_arquivo(nome_arquivo, df_tratado=''):
    df_tratado.to_excel(f'dados/{nome_arquivo}.xlsx'
)
```

```python
df = ler_arquivo('base_entregas_brutas')
```

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">

        Por segurança, criamos uma copia do DF       
    </strong>
</p>

```python
df_tratado = df.copy()
```

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Filial 
    </strong>
</p>

```python
df_tratado.loc[:, 'filial'] = df_tratado['filial'].fillna('Não informada') # : vai do inicio ao fim  -> fillna substitui dados vazios pelo texto informado

df_tratado['filial'].sample(10)
```

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Cliente 
    </strong>
</p>

```python
df_tratado.loc[:, 'cliente'] = df_tratado['cliente'].fillna('Outros')

df_tratado['cliente'].sample(10)
```

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Dias de atraso 
    </strong>
</p>

```python
df_tratado.loc[:, 'dias_atraso'] = df_tratado['dias_atraso'].fillna(99)

df_tratado['dias_atraso'].sample(10)
```

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Prazo previsto 
    </strong>
</p>

```python
df_tratado.loc[:, 'prazo_previsto'] = df_tratado['prazo_previsto'].fillna(
    df_tratado['prazo_previsto'].median()                                          #calcula a media de todos os campos preenchidos da coluna e traz o resultado
)

df_tratado
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/><br/><br/>
        Tratamento - Duplicadas
    </strong>
</p>

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Verificando duplicadas com base no ID / VERIFICANDO SE DEU CERTO
    </strong>
</p>

```python
df_tratado.duplicated(subset=['id_entrega']).sum()
```

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Removendo duplicadas, mantendo a primeira ocorrencia
    </strong>
</p>

```python
df_tratado = df_tratado.drop_duplicates(subset=['id_entrega'])                     # drop_duplicates -> deleta os duplicados
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/><br/><br/>
        Tratamento - Tipagem correta para ID
    </strong>
</p>

```python
df_tratado['id_entrega'] = df_tratado['id_entrega'].astype('Int64')                    # astype -> Altera o Tipo
df_tratado['prazo_previsto'] = df_tratado['prazo_previsto'].astype('Int64')
df_tratado['dias_atraso'] = df_tratado['dias_atraso'].astype('Int64')
```

```python
df_tratado.dtypes                                                                      # para ver o tipo do dado
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/><br/><br/>
        Tratamento - Padronização e Normalização
    </strong>
</p>

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Removendo espaços e alterando para caixa alta 
    </strong>
</p>

```python
df_tratado['status'] = df_tratado['status'].str.strip().str.upper()
df_tratado['tipo_servico'] = df_tratado['tipo_servico'].str.strip().str.upper()
```

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Status 
    </strong>
</p>

```python
mapa_status = {
    'ENTREGUE'    : 'ENTREGUE',
    'ENTREGUES'   : 'ENTREGUE',
    'EM TRANSITO' : 'EM TRÂNSITO',
    'EM TRÂNSITO' : 'EM TRÂNSITO',
    'EM TRANZITO' : 'EM TRÂNSITO',
    'ATRASADO'    : 'ATRASADO',
    'ATRAZADO'    : 'ATRASADO'
}

df_tratado['status'] = df_tratado['status'].map(mapa_status)

df_tratado.sample(5)
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/><br/><br/>
        Criando coluna calculada
    </strong>
</p>

```python
df_tratado['data_prevista'] = (
    df_tratado['data_postagem'] + 
    pd.to_timedelta(df_tratado['prazo_previsto'], unit='D')
)

df_tratado[
    [
    'data_postagem',
    'prazo_previsto',
    'data_prevista',
]
]
```

```python
df_tratado.dtypes
```

```python
hoje = pd.Timestamp.today().normalize()

df_tratado['Atraso'] = (
    hoje - df_tratado['data_prevista']
).dt.days

df_tratado['Atraso'] = df_tratado['Atraso'].clip(lower=0)


# df_tratado['Atraso'] = (
#     df_tratado['prazo_previsto'] + df_tratado['dias_atraso']
# )

df_tratado
```

    <strong><br/><br/><br/><br/><br/><br/>
        Prazo previsto de acordo com a filial

```python
df_tratado.loc[
    (df_tratado['cliente'] == "Shopee") & 
    (df_tratado['filial'] == "RJ"),
    "prazo_previsto"
] =5


df_tratado.loc[
    (df_tratado['cliente'] == "Amazon") &
    (df_tratado['filial'].isin( ["MG", "SP"] )),
     "prazo_previsto"
] =2

```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/><br/><br/>
        Atualizando arquivo para executar fórmulas
    </strong>
</p>

```python
app = xw.App(visible=False)                                            #importar atualizador sem abrir o excel

wb = xw.Book('dados/base2.xlsx')                                       #Abrir o arquivo

wb.app.calculate()                                                     #calcular o arquivo

wb.save()                                                              #salvar o arquivo

wb.close()                                                             #fechar o arquivo

app.quit()                                                             #sair do arquivo
```
