```python
dados = [
    ["COM0001", "COMBUSTIVEL",  "2710.12.59", "L", 200, "Delivery",                         5.69, "SP"],
    ["EST0001", "FILM STRETCH", "3920.10.99", "UN",100, "Operacional",                       37.50, "RJ"],
    ["GAD0001", "CANETA",       "9608.10.00", "UN", 5,  "Administrativo",                     1.50, "SP"],

]
colunas = ["Codigo", 'Descrição', 'NCM', "Tipo", "Quantidade", "Categoria","Valor", "Estado"]
df_exercicio = pd.DataFrame(dados, columns=colunas)

df_exercicio
```

```python
df_exercicio[df_exercicio['Valor'] > 30]

df_exercicio.describe()

df_exercicio['Valor'].max()


df_exercicio['Valor_Total'] = df_exercicio['Quantidade'] * df_exercicio['Valor']
df_exercicio


#df_exercicio.groupby('Estado').agg(
    #['Quantidade'].max(),
    #['Valor'].max(),
    #['Valor'].sum()
#)

df_exercicio.groupby('Estado').agg(
    maior_estoque= ('Quantidade', 'max'),
    maior_valor=('Valor', 'max'),
    soma_valor_unitario=('Valor','sum'),
    valor_estoque_total=('Valor_Total', 'sum')
)
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/>
        EXERCÍCIO 4 — Resumo por categoria
    </strong>
</p>

```python
resumo_cat = (
    df_exercicio
      .groupby('Categoria', as_index=True)
      .agg(
          itens            = ('Codigo', 'count'),        # quantidade de SKUs por categoria
          qtd_total        = ('Quantidade', 'sum'),      # soma de quantidades
          preco_medio      = ('Valor', 'mean'),          # preço médio simples
          maior_preco      = ('Valor', 'max'),           # maior preço unitário
          valor_total_cat  = ('Valor_Total', 'sum')      # soma de Quantidade × Valor
      )
      .sort_values('valor_total_cat', ascending=False)
)
resumo_cat
```
