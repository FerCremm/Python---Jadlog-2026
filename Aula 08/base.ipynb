Importar bibliotecas

```python
import pandas as pd
from openpyxl import Workbook, load_workbook
from openpyxl.styles import numbers
```

Criando dataframe

```python
df_entregas = pd.DataFrame({
    'id_entrega' : [1, 2 , 3], #chave primária
    'filial'     : ['RJ', 'MG', 'SP'],
    'status'     : ['ENTREGUE', 'ATRASADO', 'ENTREGUE']
})

df_custos = pd.DataFrame({
    'id_custo '  : [1, 2, 3],
    'id_entrega' : [1, 1, 3], 
    'custo'      : [150.89, 147.56, 278.26]
})
```

JOIN

```python
# inner join -> so retorna linha onde a conexão aconteceu

pd.merge(
    df_entregas, df_custos,
    on='id_entrega',                                                          #on= e onde conecta
    how='inner'                                                               #junta tudo e mantem o que ja tem
)
```

```python
pd.merge(
    df_entregas, df_custos,
    on='id_entrega',
    how='left'                                                             #junta a partir da tabela da esquerda
)
```

```python
pd.merge(
    df_entregas, df_custos,
    on='id_entrega',
    how='right'                                                             #junta a partir da tabela da direita
)
```

```python
pd.merge(
    df_entregas, df_custos,
    on='id_entrega',
    how='outer'                                                              #junta tudo mesmo que vazio
)
```

```python
df_entregas_idx = df_entregas.set_index('id_entrega')
df_custos_idx = df_custos.set_index('id_entrega')

df_join = df_entregas_idx.join(df_custos_idx, how='left')

df_join.reset_index()
```
