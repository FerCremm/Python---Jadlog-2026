```python

Nome_Completo = input("Nome Completo: ")                                                         # Nome do Colaborador
print("Colaborador:", Nome_Completo)

cargo = int(input("1 Analista Financeiro / 2 Assistente Financeiro / 3 - Coordenador /: "))      #Cargo do Colaborador

match cargo:
    case 1:
        print("Analista Financeiro")
    case 2:
        print("Assistente Financeiro")
    case 3: 
        print("Coordenador")
    case _:
        print("Não consta na Base")

match cargo:                                                                                      #Salario do Colaborador                                              
    case 1:
        print("R$10.000,00")
    case 2: 
        print("R$5.000,00")
    case 3:
        print("R$15.000,00")

from datetime import date

adm = input("Data de admissão (dd/mm/aaaa): ")                                                    # ex: 21/03/2022
dia, mes, ano = adm.split("/")                                                                    # separa em 3 partes

data_adm = date(int(ano), int(mes), int(dia))                                                     # cria uma data real
hoje = date.today()                                                                               # data de hoje

dias = (hoje - data_adm).days                                                                     # diferença em dias
anos = dias // 365                                                                                # anos (aprox)

print("Tempo de casa (anos):", anos)


              
```
