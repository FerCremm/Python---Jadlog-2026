Estrutura condicional IF

```python
# Condição simples

if 10 > 5:
    print("ok")
```

```python
# Condição simples - Status por avliação 

avaliação = 8.9

if avaliação >= 7.0:                          # True
    print("Mandou bem meu parceiro!")

else:                                         # Ao contrario do IF acima
    print("Vacilou em ramelão")
```

```python
# Condição composta

avaliação = 5.1

if avaliação >= 7.0:                          # Maior ou igual
    print("Mandou bem meu parceiro!")

elif avaliação >= 5.0:                        # Entre o IF e ELSE
    print("Continue tentando")

else:                                         # Ao contrario do IF acima
    print("Vacilou em ramelão")
```

# Condição Aninhada

```python
avaliação = 5.1

if avaliação >= 7.0:                          # Maior ou igual
    print("Mandou bem meu parceiro!")

elif avaliação >= 5.0:                        # Entre o IF e ELSE COM VARIABEL
    resposta = input("Tem certeza Doidão? [S ou N]: ").upper

    if resposta == "S":                       #VARIAVEL DO SIM
        print("Solta o verbo: ")
    
    elif resposta == "N":                     #VARIAVEL DO NÃO
        print("Não ta afim né...")
    
    else:                                     #VARIVAL DO VACILO ONDE O FULANO NÃO COLOCOU S OU N
        print("Não entendi mano, resposta invalida")

else:                                         # Ao contrario do IF acima
    print("Vacilou em ramelão")
```

Match/Case

```python
menu = int(input("1ºCadastrar / 2ºVisualizar / Opção: " ))             #E UMA RESPOSTA DIRETA ACIMA DA VARIAVEL

match menu:

    case 1:                                    #TE GERA UMA RESPOSTA DA OPÇÃO ESCOLHIDA
        print("escolheu 1")

    case 2:                                    #TE GERA UMA RESPOSTA DA OPÇÃO ESCOLHIDA
        print("escolheu 2")

    case _:                                    #TE GERA UMA RESPOSTA DIFENTE DAS OPÇÕES DISPONIVEIS
        print('Resposta inválida!')
            
    
```
