```python
# Entrada de dados

nome = input('Nome: ').strip().title()                                                        #Title() = arruma o Texto e primeira palavra vem com a primeira letra maiuscula   ->  #obs: todo input e um texto (utilizar somente para variar)
cargo_opcao = int(input("Cargo (1. Analista Financeiro | 2. Assistente Financeiro | 3. Coordenador): "))                                            
tempo_casa = int(input("Tempo de casa: "))                                                    #int - numero
        
```

```python
# Identificando o cargo escolhido 

match cargo_opcao:
    case 1:
        cargo = "Analista Financeiro"
        salario = 10000.0
    case 2:
        cargo = "Assistente Financeiro"
        salario = 5000.0
    case 3:
        cargo = "Coordenador"
        salario = 15000.0
    case _:
        cargo = "N√£o consta na base"
        salario = 0.0
```

```python
# Classificando senioridade (if)

if tempo_casa <= 2 and tempo_casa >= 0:
    senioridade = "Junior"

elif tempo_casa <= 5 and tempo_casa >= 3:
    senioridade = "Pleno"

elif tempo_casa >= 6 and tempo_casa <= 40:
    senioridade = "Senior"

else:
    senioridade = "Analisar senioridade"
```

```python
# Aplica√ß√£o de l√≥gica de decis√£o 

if cargo_opcao == 1 and senioridade in ("Junior", "Pleno", "Senior"):
    permissao = "Gerenciamento liberado"

elif cargo_opcao == 2 and senioridade == "Junior":
    permissao = "Apenas opera√ß√µes b√°sicas"

elif cargo_opcao == 3 and senioridade in ("Pleno", "Senior"):
    permissao = "Acesso total"

elif senioridade == "Analisar senioridade":
    permissao = "Por favor analisar"

else:
    permissao = "Permiss√£o negada"
```
